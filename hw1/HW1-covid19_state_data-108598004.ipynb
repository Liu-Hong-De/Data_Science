# Step 6 Use matplotlib show 2D images about data.
[Matplotlib](https://matplotlib.org/)
### Plot the distribution of two class (1. GDP < 58000, 2. GDP >= 58000) of COVID-19 State Data using different colors and different  marker where x-axis is the Pollution and y-axis the Mortality-rate.

```python
import matplotlib.pyplot as plt

dataset_add = dataset
dataset_add['Mortality-rate'] = dataset_add['Deaths'] / dataset_add['Infected']

dataset_small_GDP = dataset_add[(dataset_add["GDP"] < 58000)]
dataset_large_GDP = dataset_add[(dataset_add["GDP"] >= 58000)]

plt.style.use('ggplot')
plt.scatter(dataset_small_GDP['Pollution'], dataset_small_GDP['Mortality-rate'], color='c', marker='D', label='GDP < 58000')
plt.scatter(dataset_large_GDP['Pollution'], dataset_large_GDP['Mortality-rate'], color='m', marker='*', label='GDP >= 58000')
plt.xlabel('Pollution')
plt.ylabel('Mortality-rate')
plt.legend()
plt.show()
```

### Show the proportion of three class of COVID-19 State Data using pie chart.

**<font color = blue size=3> About the class: </font>** <p>
 1ã€ Mortality-rate < 0.02 <p>

```python
label = ['Mortality-rate < 0.02', 'Mortality-rate between 0.02 and 0.03', 'Mortality-rate > 0.03']
data = [len(dataset_add.loc[dataset_add['Mortality-rate']<0.02]), len(dataset_add.loc[(dataset_add['Mortality-rate']>=0.02) & (dataset_add['Mortality-rate']<=0.03)]), len(dataset_add.loc[dataset_add['Mortality-rate']>0.03])]
color = ['#BE77FF', '#ffdeae', '#87cefa']
plt.pie(data, colors=color, labels=label, autopct = "%0.3f%%", radius=3)
plt.axis('equal')
plt.show()
```
